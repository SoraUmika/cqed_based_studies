## Summary

The saved study outputs support three main conclusions.

1. A wide Gaussian envelope is the best closed-system selective family for both sideband modes.
2. A square pulse is the best fastest unselective family for both sideband modes.
3. Once the matched local transmon-coherence reference is included, no family continues to satisfy the original selective thresholds, and only the storage fast-square pulse remains threshold-valid.
4. The simultaneous two-tone extension shows that the single-photon `|g,0,1> -> |g,1,0>` transfer works nearly perfectly in the closed system through a resonant bright-state mechanism, with about `29.5 ns` transfer time at equal `12 MHz` leg couplings.
5. A detuned Raman-like two-tone drive lowers the `|f>` occupancy to about `0.12`, but stretches the transfer to about `348 ns`, where readout decay reduces the noisy peak readout occupancy to about `0.56`.
6. The same constant two-tone drive does not stay confined to the intended single-transfer ladder for `n > 1`, so the clean near-unit transfer result is specific to the single-photon case.

The selected-case variables at the top of the notebook let you jump directly to one manifold and one winner class without rerunning the full study.
