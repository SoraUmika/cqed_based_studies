## Load Saved Results (default)
This is the fast path. It reads the stored study summary and reproduces the main comparison plot.


```python
rows = summary['cases']
plt.figure(figsize=(7,4))
for n_active in config['active_windows']:
    subset = [row for row in rows if row['n_active'] == n_active]
    subset = sorted(subset, key=lambda row: row['chi_t_over_2pi'])
    xs = [row['chi_t_over_2pi'] for row in subset]
    ys = [row['unitary_opt_process_fidelity'] for row in subset]
    plt.plot(xs, ys, marker='o', label=f'N={n_active}')
plt.xlabel(r'$|\\chi| T / 2\\pi$')
plt.ylabel('Weighted process fidelity')
plt.grid(alpha=0.25)
plt.legend()
plt.show()
```
