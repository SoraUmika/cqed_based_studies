# Native Multitone SQR Fixed Multi-Input Validation

This notebook reproduces the saved cross-study multi-input validation outputs for the four recent native multitone SQR studies.


```python
from pathlib import Path
import json, csv

study_root = Path.cwd().resolve().parents[0] if Path.cwd().name == 'scripts' else Path('studies/native_multitone_sqr_fixed_multiinput_validation').resolve()
data_dir = study_root / 'data'
figures_dir = study_root / 'figures'
print('study_root =', study_root)
print('data_dir =', data_dir)
```

## User-Tunable Parameters

These parameters control which saved case summaries are displayed and which probe tiers are emphasized.


```python
config = {
    'preferred_tier': 'spanning_quartet',
    'probe_tiers': ['single_ground', 'selected_pair', 'spanning_quartet'],
    'case_labels': [
        'Residual-Z study / baseline multitone',
        'Arbitrary-rotation study / direct multitone',
        'Ideal-SQR study / direct multitone',
        'Corrected-metric study / unitary-optimized',
    ],
}
config
```

## Derived Objects

Load the saved study summary and per-probe CSV. Re-running this cell after changing the parameters above is enough for the display cells below.


```python
# --- Load saved results (default) ---
summary = json.loads((data_dir / 'study_summary.json').read_text())
with (data_dir / 'probe_state_results.csv').open() as handle:
    probe_rows = list(csv.DictReader(handle))
len(summary['cases']), len(probe_rows)

# --- Re-run with current parameters ---
# import subprocess, sys
# subprocess.run([sys.executable, str(study_root / 'scripts' / 'run_study.py')], check=True)
```

## Saved Tier Metrics

Display the key reduced target-state metrics for each representative case.


```python
for case in summary['cases']:
    tier = case['tier_summary'][config['preferred_tier']]
    print(case['plot_label'].replace('\n', ' / '))
    print('  fixed process fidelity =', round(case['fixed_process_fidelity'], 6))
    print('  preferred tier mean   =', round(tier['mean_state_fidelity'], 6))
    print('  preferred tier min    =', round(tier['min_state_fidelity'], 6))
    print('')
```
