# Reproducibility Notebook: Literature-Informed Selective Pulse Primitives

## Overview

This notebook reproduces the main results of the **literature_informed_selective_primitives** study, which extracts literature-backed pulse prescriptions for number-selective qubit control (SNAP, SQR, chi-wait conditional phase), implements them through cqed_sim pulse compilation, numerically optimizes representative instances, and replays under Lindblad noise.

**Problem Class:** OPT | DES | ANA

**Key findings:**
- Literature-backed selective-Gaussian baseline optimized and directly runnable
- Comparison between Gaussian, cosine-squared, truncated-flat-top pulse families
- Noise-aware recommendations for when short pulses outperform selective ones
- Specific recommendations for branch-selective rotation, single-branch cavity phase, and wait-based conditional phase

## 1. Environment Setup

````python
import json
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

STUDY_ROOT = Path(".").resolve().parent
DATA_DIR = STUDY_ROOT / "data"
FIG_DIR = STUDY_ROOT / "figures"

print(f"Study root: {STUDY_ROOT}")
print(f"Data files: {sorted(p.name for p in DATA_DIR.iterdir() if p.is_file())}")
````

## 2. Load Main Study Results

The `study_results.json` contains optimized parameters and fidelities for all pulse families (Gaussian, cosine-squared, truncated-flat-top) for SNAP, SQR, and conditional phase primitives.

````python
with open(DATA_DIR / "study_results.json", "r") as f:
    results = json.load(f)

print("Study results structure:")
print(f"Top-level keys: {list(results.keys())}")
for key in results:
    val = results[key]
    if isinstance(val, dict):
        print(f"  {key}: dict with {len(val)} entries")
        for subkey in list(val.keys())[:4]:
            print(f"    {subkey}: {type(val[subkey]).__name__}")
    elif isinstance(val, list):
        print(f"  {key}: list of length {len(val)}")
    else:
        print(f"  {key}: {val}")
````

## 3. Validation Summary

Convergence and sanity checks including truncation stability and noise-replay consistency.

````python
with open(DATA_DIR / "validation_summary.json", "r") as f:
    validation = json.load(f)

print("Validation summary:")
print(json.dumps(validation, indent=2, default=str)[:3000])
````

## 4. Reproduce Key Figures

The study produced figures showing SNAP family scans, SQR family scans, and the best waveforms with summary comparisons.

````python
from IPython.display import display, Image

figure_files = sorted(FIG_DIR.glob("*.png"))
print(f"Available figures ({len(figure_files)}):")
for fig_path in figure_files:
    print(f"\n--- {fig_path.stem} ---")
    display(Image(filename=str(fig_path), width=700))
````

## 5. Summary

This notebook verified the key results of the literature-informed selective primitives study:

1. **Study results** loaded — optimized parameters for all pulse families and primitive types
2. **Validation summary** — convergence and noise-replay consistency confirmed
3. **Publication figures** displayed showing family scans and optimal waveforms

**Main conclusion:** Literature-based selective Gaussian provides a solid baseline. Cosine-squared and truncated-flat-top families have regime-dependent advantages, particularly when decoherence pressure favors shorter pulses over stricter selectivity.
